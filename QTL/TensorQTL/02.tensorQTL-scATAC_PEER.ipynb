```r
library(peer)
library(data.table)
```

```r
PC <- read.csv('/media/AnalysisDisk2/Yangshichen/0_HIV_RNA/QTL/01.Dynamic/01.Data/01.Genotype/PCA/eigenvec.csv',header = T)
colnames(PC)[1] <- 'sample'
rownames(PC) <- PC[,1]
PC <- PC[,-c(1,2)]
PC
```

```r
cova_mat <- read.csv('/media/AnalysisDisk2/Yangshichen/0_HIV_RNA/QTL/01.Dynamic/01.Data/sample_information.csv',header = T)
cova_mat <- cova_mat[match(rownames(PC), cova_mat$sample), ]
rownames(cova_mat) <- cova_mat[,1]
cova_mat <- cova_mat[,-1]
cova_mat$stage <- as.numeric(factor(cova_mat$stage, levels = c("HDs", "IRs", "INRs"))) - 1
cova_mat
```

```r
meatdata <- cbind(cova_mat,PC)
meatdata <- meatdata[,c(2,3,4,5,6)]
meatdata
```

```r
# 设置目标文件夹路径
folder_path <- "/media/AnalysisDisk2/Yangshichen/0_HIV_RNA/QTL/01.Dynamic/01.Data/03.scATAC-seq/01.pseudobulk/04.top2000_normal_dis/"

# 获取所有 csv 文件的完整路径
csv_files <- list.files(path = folder_path, pattern = "\\.csv$", full.names = TRUE)
csv_files
```

```r
for (file in csv_files) {
    filename <- basename(file)
    celltype <- tools::file_path_sans_ext(filename)
    
    expr <- read.csv(file,header = T)
    colnames(expr)[1] <- 'sample'
    rownames(expr) <- expr[,1]
    expr <- expr[,-1]

    #样本匹配
    meta <- meatdata[match(rownames(expr), rownames(meatdata)), ]
    
    #Set PEER paramaters based on the instructions from PEER package website
    model = PEER()
    PEER_setPhenoMean(model, as.matrix(expr))
    PEER_setCovariates(model, as.matrix(meta))
    dim(PEER_getPhenoMean(model))

    #Set to generate 15 PEER factors: K=15
    K <- 15
    PEER_setNk(model,K)
    PEER_getNk(model)

    PEER_setNmax_iterations(model, 1000)

    #Calculate
    PEER_update(model)

    #Observing output
    pf.names <- paste('PF',1:K, sep = '')
    
    # factors: NxK matrix
    f <- PEER_getX(model)
    f_df <- data.frame(f)
    colnames(f_df) <- c(colnames(meatdata),pf.names)
    rownames(f_df) <- rownames(expr)
    write.csv(f_df, paste0("/media/AnalysisDisk2/Yangshichen/0_HIV_RNA/QTL/Data/scATAC-seq/05.PEER/01.factor/",celltype,".csv"))

    # weights: GxK matrix
    w <- PEER_getW(model)
    w_df <- data.frame(w)
    colnames(w_df) <- c(colnames(meatdata),pf.names)
    w_df$geneid <- colnames(expr)
    rownames(w_df) <- w_df$geneid
    w_df <- w_df[,-c(length(colnames(w_df)))]
    write.csv(w_df, paste0("/media/AnalysisDisk2/Yangshichen/0_HIV_RNA/QTL/Data/scATAC-seq/05.PEER/02.weights/",celltype,".csv"))
    
    # precision: Kx1 matrix
    p <- PEER_getAlpha(model)
    p_df <- data.frame(p)
    rownames(p_df) <- c(colnames(meatdata),pf.names)
    write.csv(p_df, paste0("/media/AnalysisDisk2/Yangshichen/0_HIV_RNA/QTL/Data/scATAC-seq/05.PEER/03.precision/",celltype,".csv"))

    # residuals: NxG matrix
    r <- PEER_getResiduals(model)
    r_df <- data.frame(r)
    colnames(r_df) <- colnames(expr)
    rownames(r_df) <- rownames(expr)
    write.csv(r_df, paste0("/media/AnalysisDisk2/Yangshichen/0_HIV_RNA/QTL/Data/scATAC-seq/05.PEER/04.residuals/",celltype,".csv"))
    print(paste0(celltype," calculate over!"))
}
```

### 读取数据

```r
folder_path <- "/media/AnalysisDisk1/Yanshichen/0_HIV_RNA/cpdb/数据处理L3/interacting_pair/"
file_list <- list.files(path = folder_path, pattern = "*.csv", full.names = TRUE)
```

```r
for (file_path in file_list) {
    data <- read.csv(file_path)
    data$pvalues <- ifelse(data$pvalues == 0, 0.001, data$pvalues)
    data$stage <- factor(data$stage,lev= c("HDs","IRs","INRs"))
    num_rows <- nrow(data)
    num_cols <- ncol(data)
    pdf_width <- num_cols * 1.3
    pdf_height <- num_rows * 0.1
    p = ggplot(data,aes(stage,celltype),label_format=200)
    p = p + geom_point(aes(color= means,size= -1*log10(pvalues)))
    colors <- c("#2166AC","#4393C3","#92C5DE","#D1E5F0","#FDDBC7","#F4A582","#D6604D","#B2182B")
    p = p + scale_colour_gradientn(colors = colors ,na.value="grey")
    p = p + labs(size=expression(-log[10](pvalue)),color="Interaction Score",x="",y="",title=paste0(unique(data$interacting_pair)))
    p = p + theme(legend.position = "right") 
    p = p + theme_bw() 
    p = p + theme(plot.title = element_text(hjust = 0.5)) 
    p = p + guides(shape = guide_legend(override.aes = list(size = 5)),
                   color = guide_colorbar(order = 0),fill = guide_legend(order = 0))
    ggsave(paste0("/media/AnalysisDisk1/Yanshichen/0_HIV_RNA/cpdb/数据处理L3/figure/",unique(data$interacting_pair),".pdf"),p,width = pdf_width, height = pdf_height,limitsize = FALSE)
}
```
